```python
import numpy as np # numpy 를 np라는 이름으로 사용하겠다.
import matplotlib.pyplot as plt # matplotlib.pyplot을 plt으로 정의해 사용하겠다.

# 데이터 준비
x = np.arange(0,10,0.1) # 0부터 10까지 0.1씩 증가 0,0.1,0.2~~~10
y = np.sin(x) # sin 삼각함수 사인 값을 반환

# 그래프 그리기

plt.plot(x,y) # x,y로 그래프
plt.show() # 그래프 보여주는 함수
```

```python
import numpy as np # numpy 를 np라는 이름으로 사용하겠다.
import matplotlib.pyplot as plt # matplotlib.pyplot을 plt으로 정의해 사용하겠다.


# 데이터 준비
x = np.arange(0,10,0.2) #  0.2로 증가 변경
y1 = np.sin(x) # 사인
y2 = np.cos(x) # 코사인

# 그래프 그리기
plt.plot(x, y1, label = "sin")
plt.plot(x, y2, linestyle ="--", label="cos") # 코사인 사용하고 점선으로 그리기
plt.xlabel("x") # x축 이름
plt.ylabel("y") #  y축 이름
plt.title('sin & cos use') # 제목
plt.legend()
plt.show()
```
