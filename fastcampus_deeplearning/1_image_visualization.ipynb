```python
# 이미지 시각화

# package 불러오기

import numpy as np
from PIL import Image # image 읽어올때 PIL

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# 이미지 파일 열기

path = 'image/animal1.jpg' # 다람쥐 사진 경로

image_pil = Image.open(path)
image  = np.array(image_pil)
```

```python

# 이미지 확인해보기
print(image.shape) # 201,251, 3 201 - 세로 길이, 251 - 가로, 3 - 색상 rgb, 이미지는 2차원이지만 rgb때문에 3

print(np.min(image), np.max(image)) # 0,255 확인  range확인 
```

```python
# 그래프로 시각화 하기

plt.hist(image.ravel(), 256, [0,256])
plt.show()
```

```python
# 그림 나타내기
plt.imshow(image)
plt.show()
```

```python
# 이미지 흑백으로 열기
print("path : ",path)

image_pil = Image.open(path).convert("L") # 흑백으로
image_bw = np.array(image_pil)

print("image_bw : ", image_bw.shape) # 201, 251 , rgb가 빠져서 0,1 혹은 없음
```

```python
# 흑백 이미지 열기

plt.subplot(1,2,1) # 수치적으로 점점 초록색 - 높은 값이 노랑색쪽 밝아진다.
plt.imshow(image_bw)

plt.subplot(1,2,2)
plt.imshow(image_bw, 'gray') # 흑백

plt.show()
```

```python
# 다른 색상으로 cmap 표현
# gray scale

print("image_bw : ",image_bw.shape)
print("image : ",image.shape)


# RdBu red,blue
# jet
plt.subplot(1,2,1)
plt.imshow(image_bw, 'RdBu')

plt.subplot(1,2,2)
plt.imshow(image_bw, 'jet') # 높을수록 빨간색, 낮을 수록 파란색

plt.show()

```

```python
# Colorbar 추가하기

plt.imshow(image_bw, 'jet') 
plt.colorbar() # 컬러바로 색상 설명
plt.show()
```

```python
# 이미지 사이즈 조절

plt.figure(figsize=(10,7))

plt.subplot(1,2,1)
plt.imshow(image_bw, 'jet')
plt.colorbar()

plt.subplot(1,2,2)
plt.imshow(image_bw, 'RdBu')

plt.show()
```

```python
# 이미지 제목 추가

plt.title('squirrel')
plt.imshow(image)
plt.show()
```
